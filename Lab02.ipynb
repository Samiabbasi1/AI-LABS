```python
#Task1

import random

class secsystem:
    def __init__(self):
        self.components = {'A': random.choice(['safe', 'vulnerable']),
                           'B': random.choice(['safe', 'vulnerable']),
                           'C': random.choice(['safe', 'vulnerable']),
                           'D': random.choice(['safe', 'vulnerable']),
                           'E': random.choice(['safe', 'vulnerable']),
                           'F': random.choice(['safe', 'vulnerable']),
                           'G': random.choice(['safe', 'vulnerable']),
                           'H': random.choice(['safe', 'vulnerable']),
                           'I': random.choice(['safe', 'vulnerable'])}
    def display_system(self):
        print("system state: ")
        for component, status in self.components.items():
            print(f"component {component}: {status}")
        print()
    def scan_system(self):
        print("scanning")
        vulnerable_components = []
        for component, status in self.components.items():
            if status == 'vulnerable':
                print(f"component {component} is vulnerable.")
                vulnerable_components.append(component)
            else:
                print(f"component {component} is safe.")
        return vulnerable_components
    def patch_vulnerabilities(self, vulnerable_components):
        for component in vulnerable_components:
            self.components[component] = 'safe'
            print(f"patching component {component}: now safe.")
        print()
    def final_check(self):
        print("final check:")
        self.display_system()
        print("all vulnerabilities patched.")
def cybersecurity_exercise():
    system = secsystem()
    print("initial check:")
    system.display_system()
    vulnerable_components = system.scan_system()
    if vulnerable_components:
        system.patch_vulnerabilities(vulnerable_components)
    else:
        print("no vulnerabilities detected. All components are safe.\n")
    system.final_check()
cybersecurity_exercise()

#Task2
import random
LOAD_STATES = ['Underloaded', 'Balanced', 'Overloaded']
def initialize_system():
    servers = ['Server A', 'Server B', 'Server C', 'Server D', 'Server E']
    system_state = {server: random.choice(LOAD_STATES) for server in servers}
    return system_state
def scan_system(system_state):
    overloaded_servers = []
    underloaded_servers = []
    balanced_servers = []
    for server, state in system_state.items():
        if state == 'Underloaded':
            underloaded_servers.append(server)
        elif state == 'Overloaded':
            overloaded_servers.append(server)
        else:
            balanced_servers.append(server)
    return overloaded_servers, underloaded_servers, balanced_servers
def balance_load(system_state, overloaded_servers, underloaded_servers):
    while overloaded_servers and underloaded_servers:
        overloaded_server = overloaded_servers.pop()
        underloaded_server = underloaded_servers.pop()
        print(f"moving tasks from {overloaded_server} to {underloaded_server} to balance the load.")
        system_state[overloaded_server] = 'balanced'
        system_state[underloaded_server] = 'balanced'
def final_system_check(system_state):
    print("\nfinal System State:")
    for server, state in system_state.items():
        print(f"{server}: {state}")
def simulate_load_balancing():
    print("initial System Check:")
    system_state = initialize_system()
    for server, state in system_state.items():
        print(f"{server}: {state}")
    overloaded_servers, underloaded_servers, balanced_servers = scan_system(system_state)
    print("\nOverloaded Servers:", overloaded_servers)
    print("Underloaded Servers:", underloaded_servers)
    print("\nBalancing the load...")
    balance_load(system_state, overloaded_servers, underloaded_servers)
    final_system_check(system_state)
simulate_load_balancing()



#Task3
import random
class backupmanagement:
    def __init__(self):
        self.tasks = {
            f"backup {i}": random.choice(["completed", "failed"]) for i in range(1, 11)
        }
    def displaytasks(self):
        print("task status:")
        for task, status in self.tasks.items():
            print(f"{task}: {status}")
        print()
    def scan_and_retry(self):
        print("scanning for failed backups:")
        failed_tasks = [task for task, status in self.tasks.items() if status == "failed"]
        if failed_tasks:
            print(f"failed tasks detected: {', '.join(failed_tasks)}")
            for task in failed_tasks:
                print(f"retrying {task}...")
                self.tasks[task] = "completed"
                print(f"{task} is now completed.")
        else:
            print("no failed tasks found.")
        print()
def backup_management_exercise():
    agent = backupmanagement()
    print("initial Task Statuses:")
    agent.displaytasks()
    agent.scan_and_retry()
    print("updated Task Statuses:")
    agent.displaytasks()

backup_management_exercise()

#Task4
import random
VULNERABILITY_STATES = ['Safe', 'Low Risk', 'High Risk']
def initialize_system():
    components = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
    system_state = {component: random.choice(VULNERABILITY_STATES) for component in components}
    return system_state
def scan_system(system_state):
    warnings = []
    success_messages = []
    high_risk_components = []
    low_risk_components = []
    for component, state in system_state.items():
        if state == 'Safe':
            success_messages.append(f"component {component} is safe.")
        elif state == 'Low Risk':
            warnings.append(f"component {component} has a low risk vulnerability.")
            low_risk_components.append(component)
        elif state == 'High Risk':
            warnings.append(f"component {component} has a high risk vulnerability.")
            high_risk_components.append(component)
    return warnings, success_messages, low_risk_components, high_risk_components
def patch_vulnerabilities(system_state, low_risk_components):
    for component in low_risk_components:
        system_state[component] = 'Safe'
        print(f"component {component} is now Safe.")
def final_system_check(system_state, high_risk_components):
    print("\nFinal System State:")
    for component, state in system_state.items():
        print(f"Component {component}: {state}")
    if high_risk_components:
        print("\nHigh Risk Vulnerabilities remain unresolved.Premium service required.")
def simulate_cybersecurity_exercise():
    print("initial system check:")
    system_state = initialize_system()
    for component, state in system_state.items():
        print(f"Component {component}: {state}")

    print("\nsystem scan:")
    warnings, success_messages, low_risk_components, high_risk_components = scan_system(system_state)
    for warning in warnings:
        print(warning)
    for message in success_messages:
        print(message)
    print("\npatching vulnerabilities:")
    patch_vulnerabilities(system_state, low_risk_components)
    final_system_check(system_state, high_risk_components)
simulate_cybersecurity_exercise()
```
