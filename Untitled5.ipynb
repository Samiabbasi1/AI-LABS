```python
import random
class system:
  def __init__(self):
    self.components = {
        'A':random.choice(['Safe', 'Vulnerable']),
        'B':random.choice(['Safe', 'Vulnerable']),
        'C':random.choice(['Safe', 'Vulnerable']),
        'D':random.choice(['Safe', 'Vulnerable']),
        'E':random.choice(['Safe', 'Vulnerable']),
        'F':random.choice(['Safe', 'Vulnerable']),
        'G':random.choice(['Safe', 'Vulnerable']),
        'H':random.choice(['Safe', 'Vulnerable']),
        'I':random.choice(['Safe', 'Vulnerable'])
    }

  def displaysystem(self):
    for component,status in self.components.items():
      print(f"component {component}:{status}")
      print()

  def scansystem(self):
    vulnerable = []
    for component,status in self.components.items():
      if status == 'Vulnerable':
        print(f"component {component} is vulnerable")
        vulnerable.append(component)
      else:
        print(f"component {component} is safe")
    return vulnerable


  def patch(self, vulnerable):
      for component in vulnerable:
        self.components[component] = 'Safe'
      print(f"component {component} has been patched")


  def finalcheck(self):
      self.displaysystem()
      print("all vulnerablilities have been patched")

def cyberexer():
    sys = system()
    sys.displaysystem()
    vulnerable = sys.scansystem()
    if vulnerable:
      sys.patch(vulnerable)
      sys.displaysystem()
    else:
      print("no vulnerablilities found")
    sys.finalcheck()

cyberexer()
```

```python
import random

load_states = ['Underloaded', 'Balanced', 'Overloaded']

class DataCenter:
    def __init__(self):
        self.servers = {
            'Server A': random.choice(load_states),
            'Server B': random.choice(load_states),
            'Server C': random.choice(load_states),
            'Server D': random.choice(load_states),
            'Server E': random.choice(load_states)
        }

    def display_servers(self):
        print("\nCurrent Server States:")
        for server, state in self.servers.items():
            print(f"{server}: {state}")
        print()

    def scan(self):
        underloaded, overloaded, balanced = [], [], []
        for server, state in self.servers.items():
            if state == 'Underloaded':
                underloaded.append(server)
            elif state == 'Overloaded':
                overloaded.append(server)
            else:
                balanced.append(server)
        return underloaded, overloaded, balanced

    def patch(self, underloaded, overloaded):
        while underloaded and overloaded:
            overloaded_ser = overloaded.pop()
            underloaded_ser = underloaded.pop()
            print(f"Moving load from {overloaded_ser} to {underloaded_ser}")
            self.servers[overloaded_ser] = 'Balanced'
            self.servers[underloaded_ser] = 'Balanced'
        print("\nLoad Balancing Complete!\n")

    def finalcheck(self):
        print("\nFinal Server Load States:")
        self.display_servers()


def run_datacenter():
    dc = DataCenter()
    dc.display_servers()

    underloaded, overloaded, balanced = dc.scan()

    if underloaded and overloaded:
        dc.patch(underloaded, overloaded)
        dc.display_servers()
    else:
        print("No load balancing required.\n")

    dc.finalcheck()


run_datacenter()

```

```python
  import random
  class backup:
    def __init__(self):
      self.tasks = {f"backup{i}":random.choice(["Completed","Failed"]) for i in range(1,11)}

    def display(self):
      for task,status in self.tasks.items():
        print(f"{task}:{status}")

    def scan(self):
      failed = []
      for task,status in self.tasks.items():
        if status == "Failed":
          failed.append(task)
      if failed:
        for task in failed:
          print(f"retrying {task}")
          self.tasks[task] = "Completed"
        self.display()
      else:
        print("no failed tasks")



  def backupmanage():
    agent = backup()
    agent.display()
    agent.scan()
    agent.display()

  backupmanage()
```

```python
import random

vulnerable_states = ['Safe', 'High Risk', 'Low Risk']

class Vulnerability:
    def __init__(self):
        self.components = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']
        self.sysstate = {component: random.choice(vulnerable_states) for component in self.components}

    def display(self):
        print("\nSystem State:")
        for component, state in self.sysstate.items():
            print(f"{component}: {state}")
        print()

    def scansys(self):
        warnings = []
        success = []
        high_risk = []
        low_risk = []

        for component, state in self.sysstate.items():
            if state == 'Safe':
                success.append(f"{component} is safe")
            elif state == 'Low Risk':
                low_risk.append(component)
            elif state == 'High Risk':
                warnings.append(f"WARNING: {component} is at HIGH RISK!")
                high_risk.append(component)

        return success, warnings, low_risk, high_risk

    def patch(self, low_risk):
        for component in low_risk:
            self.sysstate[component] = 'Safe'
        print("\nLow-risk vulnerabilities patched!\n")

    def finalcheck(self, high_risk):
        self.display()
        if high_risk:
            print(" Premium Security Service Required! High-risk vulnerabilities detected.\n")
        else:
            print(" All vulnerabilities have been resolved. System is secure!\n")

def runsys():
    sys = Vulnerability()
    sys.display()

    success, warnings, low_risk, high_risk = sys.scansys()

    for warning in warnings:
        print(warning)
    for message in success:
        print(message)

    if low_risk:
        print("\n Patching vulnerabilities...\n")
        sys.patch(low_risk)

    sys.finalcheck(high_risk)

runsys()
```

```python
import time
import random

class Hospitalrobot:
    def __init__(self):
        self.locations = ["Storage", "Corridor", "Nurse Station", "Room 101", "Room 102", "Room 103"]
        self.patients = {
            "Room 101": {"medicine": "Painkiller", "id": "P101"},
            "Room 102": {"medicine": "Antibiotic", "id": "P102"},
            "Room 103": {"medicine": "Vitamin", "id": "P103"}
        }
        self.robotloc = "Storage"
        self.medicine = None

    def move(self, location):
        print(f" Moving to {location}.")
        self.robotloc = location

    def pickup(self, room):
        if self.robotloc == "Storage":
            self.medicine = self.patients[room]["medicine"]
            print(f"Picked up {self.medicine} for {room}.")
        else:
            print(" Must be in Storage to pick up medicine!")

    def deliver(self, room):
        if self.robotloc == room and self.medicine == self.patients[room]["medicine"]:
            print(f" Scanning Patient ID in {room}.")
            if self.scan_patient_id(room):
                print(f"Delivered {self.medicine} to {room} successfully\n")
                self.medicine = None
            else:
                print(f"Incorrect ID! Alerting staff for {room}.\n")
                self.alert_staff(room)
        else:
            print(f"Cannot deliver to {room}.\n")

    def scan_patient_id(self, room):
        return random.choice([True, False])

    def alert_staff(self, room):
        print(f"Alerting nurse for assistance in {room}\n")

    def execute_delivery(self):
        for room in self.patients:
            self.move("Storage")
            self.pickup(room)
            self.move(room)
            self.deliver(room)

robot = Hospitalrobot()
robot.execute_delivery()
```

```python
class FirefightingRobot:
    def __init__(self):
        self.grid = {
            'a': 'Safe', 'b': 'Safe', 'c': 'Fire',
            'd': 'Safe', 'e': 'Fire', 'f': 'Safe',
            'g': 'Safe', 'h': 'Safe', 'j': 'Fire'
        }
        self.path = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'j']
        self.position = 'a'

    def display_grid(self):
        print("\nCurrent Building Status:")
        print(f"{self.grid['a']} | {self.grid['b']} | {self.grid['c']}")
        print(f"{self.grid['d']} | {self.grid['e']} | {self.grid['f']}")
        print(f"{self.grid['g']} | {self.grid['h']} | {self.grid['j']}")
        print("-" * 30)

    def move_and_extinguish(self):
        for room in self.path:
            self.position = room
            print(f"Moving to room {room}")

            if self.grid[room] == 'Fire':
                print(f"Fire detected in room {room}. Extinguishing...")
                self.grid[room] = 'Safe'
                print(f"Fire in room {room} extinguished.")

            self.display_grid()

        print("All fires have been extinguished.")

robot = FirefightingRobot()
robot.display_grid()
robot.move_and_extinguish()
```
